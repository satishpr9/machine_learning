```python
import torch
import torch.nn as nn
```

```python
class Model(nn.Module):
  def __init__(self,nums_features):
    super().__init__()
    self.linear=nn.Linear(nums_features,3)
    self.rule=nn.ReLU()
    self.linear1=nn.Linear(3,1)
    self.sigmoid=nn.Sigmoid()

   # self.network =nn.Sequential(
    #    nn.Linear(nums_features,3)
    #nn.ReLU()
    #nn.Linear(3,1)
    #nn.Sigmoid()
    #)
  def forward(self,features):
    out=self.linear(features)
    out=self.rule(out)
    out=self.linear1(out)
    out=self.sigmoid(out)
    return out
```

```python
#create dataset
features=torch.rand(10,6)
#create  model
model=Model(features.shape[1])
#call model for forward pass
model(features)
```
